## An example of the Wigner function for a two-mode quantum state.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from shmidt import shmidt_decomp
```

```python

def f(x, y):
    return np.exp(-(x**2 + y**2)/0.2) * (np.sin(x*y/0.1) + 1j * np.cos(x*y/0.1))

x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)
F = f(X, Y)


fig = plt.figure(figsize=(9, 11))
plt.subplot(211)
plt.contourf(X, Y, np.real(F), 100, cmap='RdGy')
plt.title('Real(F(x,y))', fontsize=16)
plt.colorbar()
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.subplot(212)
plt.contourf(X, Y, np.imag(F), 100, cmap='RdGy')
plt.title('Imag(F(x,y))', fontsize=16)
plt.colorbar()
plt.xlabel('X', fontsize=16)
plt.ylabel('Y', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Decomposition.
modes1, modes2, eigvals, phases = shmidt_decomp(F, modes_num=40)
```

```python
# Plotting modes and eigenvalues.
fig = plt.figure(figsize=(11, 9))
plt.subplot(221)
plt.plot(np.abs(modes1[0, :]))
plt.xlabel('X')
plt.title("1'st mode.")
plt.ylabel('Amplitude')
plt.subplot(222)
plt.plot(np.abs(modes1[1, :]))
plt.xlabel('X')
plt.title("2'nd mode.")
plt.ylabel('Amplitude')
plt.subplot(223)
plt.plot(np.abs(modes1[2, :]))
plt.xlabel('X')
plt.title("3'rd mode.")
plt.ylabel('Amplitude')
plt.subplot(224)
plt.plot(np.abs(modes1[3, :]))
plt.xlabel('X')
plt.title("4'th mode.")
plt.ylabel('Amplitude')
plt.show()


fig2 = plt.figure(figsize=(11.5, 6))
plt.plot(eigvals[:10])
plt.show()
```
